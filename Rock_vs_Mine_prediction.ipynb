importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Data Proessing

```python
#loading the dataset to a pandas datafram
sonar_data = pd.read_csv('/sonar data.csv', header=None)
```

```python
sonar_data.head()
```

```python
#number of rows and coloums
sonar_data.shape
```

```python
sonar_data.describe #describe --> statistical measures of the data
```

```python
sonar_data[60].value_counts()
```

M --> Mine

R --> Rock

```python
sonar_data.groupby(60).mean() #seperating rock and mine
```

```python
#separating data and labels(rock and mine)
X = sonar_data.drop(columns=60, axis=1) #storing 60 coloum in y
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

Training and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

original x (208 example),training data (187 instance) test data(21 instance)

training machine learning model with this x train data
Model Traning --> Logistic Regression

```python
print(X_train)
print(Y_train)
```

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```
